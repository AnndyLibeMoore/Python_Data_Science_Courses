# ¿Qué es Pandas?

Pandas es una de las bibliotecas más poderosad e Python para manipulación y análisis de datos. Te permite trabajar con datos en formato de tablas (DataFrame) de manera sencila y eficiente, similar a una hoja de cálculo peor con todo el poder de Python detrás. Con Pandas puedes filtrar, organizar, limpiar y transformar grandes conjuntos de datos en pocas lineas de código sin necesidad de escribir bucles o realizar tareas manuales repetitivas.\

Documentación: https://pandas.pydata.org/docs/

## 1.Estructuras de Datos

## Tipos de estructuras

### Series

Son arrays unidimensionales etiquetados donde solo tenemos un formato de una sola infomación o sea una sola columna para almacenar en esta podemos almacenar cualquier tipo de dato. Las etiquetas de las lineas se denominan **index**. \
La forma básica de crear una *Series* es: \
\

pd.Series(datos, index = index) \
\

El argumento datos puede ser un diccionario, una lista, un array, un numpy o una constante. \
\

**DataFrames**\
Es una estructura de datos Tabular bidimensional con etiquetas de Fila y Columna. Como las Series los DataFrames son capaces de almacenar cualquier tipo de datos.\
\

df = pd.DataFrame(datos, index = index, columns = columns)\\
El argumento datos puede ser un diccionario, una lista, un array, un numpy o una constante.

# 4.Iterando en DataFrames

```python
dataset.head()
```

Para iterar debemos recorrer el df y eso lo haremos con un bucle for

```python
#aqui recorremos las columnas

for item in dataset:
  print(item)
```

Digamos que quremos obtener el dato del km promedio
que han recorrido los carros, eso lo hacemos dividiendo el
km_total / (año actual - año_fabrica) paar este ejemplo el año actual es 2024
pero tenemos algunos vehiculos que so del 2024 asi que si hacemos una division
teniendo como divisor la resta de 2024 - 2024 pues daria 0 y si dividimos por 0
pues sera error asi que para evitar esto haremos un **if** \
  *if ((2024 - row['Año']) != 0):* \
   primero restamos año actual - año de fabricacion que esta en la columna año si es diferenete a 0 (!=0) entonces seguimos con el for\

*       dataset.loc[index, 'km_media'] = row['Kilometraje'] / (2024 - row['Año'])*\
crearemos una columna que guardara el resultado, entonces aqui le decimos a python, vé al dataset con el metodo .loc que es el que busca los indices, busca el indice index que es el que esta haciendo el recorrido en el for, en la columna **km_media** como esta columna no existe, pyhon la creará.
Ubicate en ese indice con el metodo .loc y en ese columna que has creado guarda el resultado de la ecuación que te voy a dar y entonces alli realiza la operación.
necesitamos kilometraje / entre el año actual y el año de fabircación, entonces debes ir a la columna kilometraje (row['kilometraje'] / resta, esta resta ya la contiene el if, que es (2024 - row['Año']).\
\
Y qué pasa si el resultado de la resta inicial es 0 pues ponemos 0 y evitamos que nos salte error y para ello hacemos el **else**\

else: \
 *dataset.loc[index, 'km_media'] = 0*\
simplemente creamos o pedimos que guarde en la misma columna que se creo esos valores y los pasamos entonces por defecto como 0




```python
#recorremos las filas con iterrows
#las filas contienen dos datos, el indice de la fila como tal
#y la posicion de cada dato que contienen las columnas o el registro completo
#con item. rows puedo acceder a todos esos datos

for index, row in dataset.iterrows():
  if ((2024 - row['Año']) != 0):  #primero restamos año actual - año de fabricacion que esta en la columna año si es diferenete a 0 (!=0) entonces seguimos con el for
    dataset.loc[index, 'km_media'] = row['Kilometraje'] / (2024 - row['Año'])#crearemos una columna que guardara el resultado
  else:
   dataset.loc[index, 'km_media'] = 0

dataset.head()
```
