### #Vemos aqui que el a√±o de mayor ventas fue el 2022, mostrando un crecimiento del 23%, auqnue  no ha sido el a√±o con  mayor porcentaje de crecimiento vemos que los n√∫meros son buenos, asi que es un buen momento para invertir con miras a crecer.

```python
a√±os = [2019, 2020, 2021, 2022]
ventas_totales = [1380596.58, 1471121.29, 1855963.32, 2288975.53]

fig, ax = plt.subplots(figsize=(8, 4))

# Calcula el porcentaje de crecimiento y almac√©nalo en una lista
porcentaje_crecimiento = [(ventas_totales[i] - ventas_totales[i-1]) / ventas_totales[i-1] * 100 for i in range(1, len(ventas_totales))]

# Agrega un gr√°fico de barras para el porcentaje de crecimiento
ax.bar(a√±os[1:], porcentaje_crecimiento, color='green', width=0.5, label='Porcentaje de Crecimiento')

ax.set_title('Porcentaje de Crecimiento de Ventas Anuales')
ax.set_xlabel('A√±o')
ax.set_ylabel('Porcentaje de Crecimiento (%)')

ax.set_xticks(a√±os)
ax.set_xticklabels([str(int(a)) for a in a√±os])  # Convierte los a√±os a enteros y los muestra como strings

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)

# Ajusta los m√°rgenes en la parte superior e inferior
ax.margins(y=0.2)

# Oculta los ticks en el eje y
ax.yaxis.set_ticks_position('none')

# A√±ade etiquetas de texto con los valores de porcentaje de crecimiento
for i, bar in enumerate(ax.patches):
    valor = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2.0, valor, f"{valor:.2f}%", ha='center', va='bottom')

descripcion = ("El a√±o con mayor crecimiento fue 2021.\n"
               "teniendo un aumento del 26,16%"
)
ax.text(1.05, 0.5, descripcion, transform=ax.transAxes, fontsize=10, va='center', ha='left',multialignment='left')
```
