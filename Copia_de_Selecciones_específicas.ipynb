Formas de Selección

```python
import pandas as pd
```

```python
#aqui tenemos una cadana de texto y el méodo split divide cada elemto en
#elementos dentro de una lista
'11 12 13 14'.split()
```

```python
#aqui creare nombre de filas y columnas con el split tambien
data = [(1, 2, 3),
        (5, 6, 7),
        (8, 10, 11, 12),
        (13, 14, 15, 16)]

df = pd.DataFrame (data, '11 12 13 14'.split(), 'c1 c2 c3 c4'.split())
df
```

```python
#veremos formas espcificas de selección, muestras y así
```

```python
#quiero ver la primera columna
df['c1']
```

```python
#recordemos que cuando es una columna crea una serie si fueran más crearia un datframe
type(df['c1'])
```

```python
#quiero ver dos columnas que estan separadas para ello debo crear una lista
#para eso pongo dos corchetes y luego paso los nombres de las columnas

df[['c3', 'c1']]
```

```python
type(df[['c3', 'c1']])
```

#forma 2

```python
#selección por indices
#quiero ver desde la fila 1 en adelante, asi que pondre en los corchetes
#posicion inicial, posición final (aqui quiero la fila 1 y 2)
#ponemos el 3 porque si porngo 2 no lo tomará n cuenta se pone la posición siguiente de la final que quiero

df[1:3]
```

```python
df
```

```python
#quiro las ultimas tres filas y algunas columnas
df[1:][['c3', 'c1']]
```

```python
df
```

```python
#ahora usaremos un méodo que hace selección ó particiones
#este método usa los nombres de las columnas aqui quiero la 13
#me muestra una serie
df.loc['13']
```

```python
#dos columnas
df.loc[['13','12']]
```

```python
df
```

```python
#quiero solo 1 fila y 1 columna
#quiero el datos que tiene la intersección de la fila 0 con la columna 1
df.loc['11','c2']
```

```python
#tenemos el método iloc
#este usa las posiciones de los indices
df.iloc[0,1]

#aqui accedo al mismo valor q en la anterior pero el método ya usa es los indices
```

```python
#ahora quiero una interseccion
#2 filas 2 columnas. Quiero una matriz
#primero las filas
df.loc[['13','11'],['c4','c1']]
```

```python
#ahora con el iloc
df.iloc[[2,0],[3,0]]
```
