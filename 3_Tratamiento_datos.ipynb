Aqui nos genera un error, porque al hacer la llmada, el servidor recibe ese pedido pero no tenemos permiso para leer esos datos porque es nuestro computador el que lee los datos. cuando accedemos coo usuarios a esas paginas estamos dando nuestros datos pero si somos una maquina, osea si queremos acceder con la llamada de colab, pues colab es una maquina y por eso no nos deja leer los datos.

**Cómo podemos hacer para leerla entonces?**
con la biblioteca *request* y con el método de la biblioteca *IO*, el método *stringIO* que nos ayuda a interpretar textos de páginas mejor.

Bien, para que funcione crearemos credenciales de usuario para que el servidor lo lea coo una oersona y no maquina, usaremos nuestro *User-Agent*.

Para obtenerlo ponemos en el buscador
*my user-agent*
en mi caso es
Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/132.0.0.0 Safari/537.36.

Bien ahora hay que enviar esta información junta para que funcioine y lo haremos con *requests.get()*
a esta requests le pasaremos la url a la que debe ir y el headers que sera mi usuario por asi decirlo.

A esta petición que haremos pues obtendremos una respuesta y esa respuesta la guardaremos en la variable *response*

Esta respuesta vendra en lenguaje nativo pero hay que transformarlo en un lenguaje q pandas pueda interpretar asi que usaremos el *(StringIO(response.text))*
pandas lo interpretara y lo transformara a un Df y lo guardara en la variable q elegimos df_html.

Recordemos que esta página tiene varias tablas, asi que podemos poner
len(df_html) y el resultado sera el tamaño de las tablas que tiene la pagina osea la cantidad de tablas, en este caso son 3.

O podemos traer tabla por tabla com ya lo habiamos hecho con
df_html[0] y así.


```python
import requests
from io import StringIO
url = 'https://www.federalreserve.gov/releases/h3/current/default.htm'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/132.0.0.0 Safari/537.36'}
response = requests.get(url, headers=headers)

df_html = pd.read_html(StringIO(response.text))
df_html[0]
```

Aún tenemos dos columnas con nulos, generalmente el cliente es quien dice si se eliminan los nulos, se reemplazan valores o cómo se actua, para ello generalmente debemos presentar un muestreo de os datos que presentan nulos, vamos a hacer esto.

Supongamos que el cliente nos dice que si en la columna Mantenimiento es nulo y es un departamento se puede eliminar pero para los demas tipos de inmuebles debemos mantenerlos.

Aqui tenemos entonces una condición par eliminar ciertos nulos, vamos a hacerlo.

```python
dataset[dataset['Mantenimiento'].isnull()].shape[0]
#total de resgistros nulos en la columna mantenimiento
```

```python
dataset[dataset['Tipo'] == 'Departamento']
```

# 5.Feature Engineering

Este proceso sirve para obtener mas información de los datos en base a lo que ya tenemos, esto ayuda por ejemplo a que en el proceso de entrenamiento el modelo entienda mejor la información y pueda llegar a mejores resultados asi como a dar un análisis de los datos mejor y más detallado brindando mejores conclusiones y permitiendo asi desarrollar mejores estrategias de implementación.

Aqui podemos ver que tenemos 3 columnas que nos dan valores separados del inmueble, estos valores aportaran al valor bruto o total del inmueble, asi que esa información podriamos obtenerla, el Valor bruto del inmueble.

Vamos entonces a obtener ese dato.

```python
#crearemos esa columna con el dato

dataset['Valor Bruto'] = dataset.Valor + dataset.Mantenimiento + dataset.Impuesto
dataset.head()
```

Ahora quiero saber el valor por metro cuadrado
esto lo hacemos con la columna valor y area

Importante que ninguna de las dos columnas esté en 0 o nulos para que no nos generé error al intentar obtner los datos, por ello se hace limpieza antes.


```python
dataset['Valor m2'] = dataset.Valor / dataset.Area
dataset.head()
```

```python
#redondeemos los valores de esta columna y sobreescribimos
dataset['Valor m2'] = dataset['Valor m2'].round(2)
dataset.head()
```

Ahora el valor bruto teniendo en cuenta el metro cuadrado, osea el valor bruto por m2

```python
dataset['Valor bruto m2'] = (dataset['Valor Bruto'] / dataset.Area).round(2)
dataset.head()
```

que hace una representación matematica generando una matriz de todos contra todos, la cual esta en corr y queremos mostrarla con colores eso lo hacemos con cmap = 'coolwarn' con lo cual le decimos que queremos que parezca un grafico de color\
los ticks los pondremos segun la cantidad de columnas que tengamos eso lo haremos con **len** que nos dice la cantidad total que tenmos en este caso de columnas, y añadiremos un **range** para que nos genere la secuencia de números que necesitamos basada en la cantidad de columnas, recordemos que aqui el dato lo tiene **corr.columns**






```python
g1.scatter(dataset_new.Valor, dataset_new.Area)
g1.set_title('Valor x Area')
g1.set_xlabel('Valor')
g1.set_ylabel('Area')

g2.hist(dataset_new.Valor)
g2.set_title('Histograma de Valor')
g2.set_xlabel('Valor')
g2.set_ylabel('Frecuencia')

grupo = dataset_new[dataset_new['Tipo'].str.contains('Casa')].groupby('Tipo')['Valor']
label =grupo.mean().index
valores = grupo.mean().values
g3.bar(label, valores)
g3.set_title('Valor Promedio por Tipo')
g3.set_xlabel('Tipo')
g3.set_ylabel('Valor')

corr = dataset_new.select_dtypes(include=['int64', 'float64']).corr()
g4.matshow(corr, cmap= 'coolwarm')
g4.set_xticks(range(len(corr.columns)))
g4.set_yticks(range(len(corr.columns)))
g4.set_xticklabels(corr.columns, fontsize= 8, rotation=90)  #rotamos para que o quede muy grande
g4.set_yticklabels(corr.columns, fontsize= 8, ) #para los nombres usamos el mismo nombre de las columnas que ya se tienen
g4.xaxis.set_ticks_position('bottom') #quiero que los ticks osea el boton aparezcan abajo osea bottom
g4.xaxis.set_label_position('bottom') # y el label que acompaña el tick o boton tambien abajo
```
