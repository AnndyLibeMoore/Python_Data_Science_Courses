## *zip()*

https://docs.python.org/3.6/library/functions.html#zip

````python
#crea un iterador de tuplas como range con listas
#tenemos dos listas
carros = ['Jetta Variant', 'Passat', 'Crossfox', 'DS5']
carros
````

````python
valores = [88078.64, 106161.94, 72832.16, 124549.07]
valores
````

````python
#las usaremos para crear un iterador con tupla,
#con nuestro metodo zip
zip(carros, valores)
 #me muetsra una confirmacion de q se hizo zip
````

````python
list(zip(carros, valores))
#quiero ver los elementos del iterador asi q los convierto en lista
#cada elemento de mi iterador es una lista, una lista q combino
#el valor de una lista con la otra
````

````python
# es una lista pero cada elemento es una tupla
#con el nombre y valor de cada vehículo
#vamos a crear un bucle

for item in zip(carros,valores):
  print(item)
  #veremos cada elemnto del iterador, cada elemento es tupla
````

````python
#desempacaremos esto, mis tuplas tienen dos valores asi que
#lo separaremos en dos variables,
#cambiamos item por las varibles donde desempacare
#q sera carro y valor

for carro, valor in zip(carros,valores):
    print(carro, valor)
````

````python
#ahora puedo trabajar con estos datos,
#pongamos una condicional
#pondremos una condicion de q valga mas de 100mil

for carro, valor in zip(carros,valores):
  if (valor > 100000 ):
     print(carro, valor)
````

````python
#si quiero que solo me muestre el nombre del carro q cumple
#con la condicion
for carro, valor in zip(carros,valores):
  if (valor > 100000 ):
     print(carro)
````

# 3.1 Creando diccionarios

Listas son colecciones secuenciales, es decir, los elementos de estas secuencias están ordenados y utilizan índices (números enteros) para acceder a los valores.

Los diccionarios son colecciones ligeramente diferentes. Son estructuras de datos que representan un tipo de mapeo. Los mapeos son colecciones de asociaciones entre pares de valores donde el primer elemento del par se conoce como llave (*key*) y el segundo como valor (*value*).

```
diccionario = {key_1: value_1, key_2: value_2, ..., key_n: value_n}
```

https://docs.python.org/3.6/library/stdtypes.html#typesmapping

````python
# [] -> lista
# () -> tupla
# {} -> diccionario

#en las tuplas y listas accedemos a sus datos conociendo la posicion del elemento
#en los diccionarios un elemento es un conjunto de dos
# la key y el value
#para acceder al value hay que conocer la key
````

# 4.3 Definiendo funciones que devuelven valores

### Funciones que devuelven un valor

#### Formato estándar

```
def <nombre>(<param_1>, <param_2>, ..., <param_n>):
    <instrucciones>
    return <resultado>
```

````python
'''
a diferencia de las  funciones anteriores
queremos q esta nos retorne un valor, para
asi poder asignar la función a una variable
esto con return
'''

def media (lista):
  valor = sum(lista) / len(lista)
  return(valor)
````

````python
result = media([1,2,3,4,5,6,7,8])
````

````python
result
````

````python
type(result)
#aqui ya retorna por tanto ya asigna un
#tipo de valor
````

### Funciones que devuelven más de un valor

#### Formato estándar

```
def <nombre>(<param_1>, <param_2>, ..., <param_n>):
    <instrucciones>
    return (<resultado_1>, <resultado_2>, ..., <resultado_n>)
```

````python
'''
bien, aqui tendremos una función q nos va
a retornar varios resultados, osea q
retornará una tupla.

devolvera el valor reultante y el tamaño
de la lista
'''

def media (lista):
  valor = sum(lista) / len(lista)
  return(valor,len(lista))
````

````python
media([1,2,3,4,5,6,7,8])
````

````python
#ahora lo asignare, cada resultado  de mi
#tupla a una variable diferente, ó a
#desempacar

resultado, n = media([1,2,3,4,5,6,7,8,9])
````

## Estructuras de Datos

### Series

Series son arrays unidimensionales etiquetados capaces de almacenar cualquier tipo de dato. Al ser unidimensionales todo se guardará secucnialmente en una sola columna. Las etiquetas de las líneas se denominan **index**. Una serie se puede crear a partir de cualquier dato, lista, array numpy, diccionario, etc, podemos tambien personalizar los indices.La forma básica de crear una Series es la siguiente:


```
    s = pd.Series(datos, index = index)
```

El argumento *datos* puede ser un diccionario, una lista, un array Numpy o una constante.

### DataFrames

DataFrame es una estructura de datos tabular bidimensional con etiquetas de fils y columnas. Como la Series, los DataFrames son capaces de almacenar cualquier tipo de datos.Para crearlo:


```
    df = pd.DataFrame(datos, index = index, columns = columns)
```

El argumento *datos* puede ser un diccionario, una lista, un array Numpy, una Series y otro DataFrame.

**Documentación:** https://pandas.pydata.org/pandas-docs/version/1.1.0/

````python
#obtuvimos una tupla asi que hay que desempacar tendre 2 variables (index y row)
#estas iran dentro del dataset.interrows q me devolvera la tupla
#haré tratamiento de datos para evitar q me salga un error pues al buscar la media
#tendre q restar años y si resto 2020 con 2020 podra darme error
#recordemos q las tuplas q me trae me dan la fila y luego la columna  el indice
#lo guardare en index sera cada columna y la fila en row, calcularemos el km
#promedio, primero haré una condicional me voy dentro de cada fila y llamo el
#nombre de la columna con la q voy a comparar en este caso Año en la condicion
# en caso de que la resta sea diferente a 0 no dara error, esto lo guardaremos en la fila
#con el loc nos ubicaremos en la fila y la columna que queremos y alli lo guardaremos
#entonces la fila en la que estoy es index, alli se guardara y
# la columna la creare y sera km_media, guardaremos en la misma fila diferente columna
#necesito el dato del km recorrido que esta en la column Kilometraje
#le digo q en la misma fila que estoy vaya a la columna kilometraje y haga la operacion para obtener la media
#esto sera si los años son diferentes
#si los años son los mismos para q no de error en esa nueva columna guardaremso el valor 0

for index, row in dataset.iterrows():
   if(2020 - row.Año != 0 ):  #aqui estoy dentro de cada fila con row
      #en caso de q la resta sea diferente de 0 no dara error
     dataset.loc[index, 'km_media'] = row.Kilometraje / (2020 - row.Año)
   #me ubico en loc donde guardare loc[fila donde estoy, columna donde guardara]
   #guadamos en lamisma fila pero nueva columna
   #el km recorrido esta en la columna km asi q voy alli
   #row.Kilometraje y hago la operacion
   #resta entre año actual y año de fabricacion
   else:
       dataset.loc[index, 'km_media'] = 0
       #esta condicion evita q halla error, si es cero la resta
       #guardaremos ese valor en la nueva columna q creamos
dataset

#obtendre el mismo dataset con una nueva columna
````
