```python
import pandas as pd
```

```python
#crearemos un data frame a partir de una lista de listas
data = [[1,2,3], [4,5,6], [7,8,9]]
data
```

```python
list('321')
```

```python
#aqui creamos una lista con caracteres de texto, una cadena de texto
#y la lista lo que hace es convertir cada caracter en un elemento de
#la lista
#esta lista sera el nombre de las lineas del df
```

```python
#recordemos q para crear el df primero decimos dónde estan los datos,
#luego le decimo el nombre que queros que tengan las lineas
#y luego el nombre de las columnas

df = pd.DataFrame(data, list('321'), list('ZXY'))
df
```

```python
#ahora veremos cómo organizarlo
df.sort_index(inplace=True)
df
```

```python
df.sort_index(inplace=True, axis = 1)  #axis organiza por columna
df
```

```python
#ahora los contenidos estan desordenados
#este método pide una lista de columnas para organizqr su contenido
#tomaremos la columna X y led remos q la organice de forma descendente
df.sort_values(by = 'X', inplace=True)
df
```

```python
#aqui se organizaron los contenidos de la columna X
#pero los indices vuelven a estar desordenados, cómo
#puedo organizar no por columna sino por fila?
df.sort_values(by = '3', axis = 1, inplace = True)
df
```

```python
#se organizaron los contenidos pero las columnas y filas quedaron
#en otro orden, esto solo para ver que se puede cambiar el orden de
#filas y columnas segun lo necesitemos
```
