```python
''' cómo pueo crear id?
par esto debemos usar Numpy'''

import numpy as np
```

```python
''' generaremos numeros aleatorios para generar el id
para generar siempre los mismos numeros, entonces le daremos
como parametro una semilla o un #seed'''

np.random.seed(123)
```

```python
total_alumnos = len(nombres)
total_alumnos
```

```python
#creamos una nueva columna
# le pasamos loq guardara
''' este id se generara aleatoriamente
lo haremos con numpy, con su metodo ramdom, pero yo quiero
que estos numeros sean en secuencia, para eso usamos el metodo
permutation est metodo nos pide el total de la secuencia, en este caso
seria el total de alumnos, q tenemos en la variable total_alumnos
y siempre debemos sumarle uno para q genere de uno en uno hasta
completar la secuencia
'''

nombres['id_alumno'] = np.random.permutation(total_alumnos) + 1
nombres.sample(5)
```

```python
'''ahora crearemos el email'''
#para ello crearemos un dominio osea el @...
#haremos una lista de dominios con 2 valores

dominios = ['@dominiodeemail.com', '@serviciodeemail.com']
```

```python
""" lo que queremos hacer es q sucesivamente uno de los dos dominios
sean asignados a cada estudiante"""

#crearemos la nueva columna
#ahora suaremos el metodo '''choice''' este elegira uno de los dos dominios
#y lo asignara a cada alumno
#a este parametro choice debo darle 2 parametros primero el parametro
# donde se encuentra la lista de la q debe elegir q en este caso
#es la lista de dominios y el segundo el total de veces que debe hacerlo
#en este caso seran 400 porq son 400 alumnos, pero este toatl
#oo tengo enla variable total_alumnos

nombres['dominio'] = np.random.choice(dominios, total_alumnos )
nombres.sample(5)
```

```python
''' ahora el usurario del correo sera el ombre del estudiante'''
#para arreglar esto solo debemos concatenar el nombre con el dominio
#crearemos una columna para el correo y alli concatenaremos

""" en este caso para concatenar llamamos al df.columna, la columna debe
estar en formato str y añadimos cat de concatenar y llamamos
de nuevo al df y la columna con la que queremos concatenar"""
#añadimos str.lower() para pedir q el resultado de esta concateación
#se vea en un string enminusculas

nombres['email'] = nombres.nombre.str.cat(nombres.dominio).str.lower()
nombres.sample(5)
```

#Matriculando Alumnos

```python
'''
relacionaremos los df q ya hemos contruido, el de alumnos y cursos
para las matriculas. Matricularemos a los alumnos en diferentes cursos, asi
cada curso tendra mas o menos alumnos
'''
```

```python
#cómo creamos una relación entre los df?
'''
tenemso el df nombres donde añadiremos una columna nueva que
muestre los cursos a los q se ha inscrito el alumno
'''

nombres
```

```python
'''
para trabjar con numeros aleatorios lo haremos con ramdon,
crearemos la columna y usaremos np para los numeros aleatorios
usaremos un metodo q me de como resultado numeros q seran exponenciales
para unos registros numeros pequeños y para otros muy gradnes, ese el
el metodo 'exponential'.
damos como parametro el tamaño ó las veces q necesitamso q se generern
esos numeros q sera el total_alumnos y pedimos q sea en formato int

'''

nombres['matriculas']  = np.random.exponential(size=total_alumnos).astype(int)
nombres.head()
```

```python
#aqui tenemos registro con 0 matriculas, asi que debemos modificarlo
#para ello redondearemos el resultado a numero superior, lo haremos
#con np y transformaremos con np.ceil asi lo transformaremos o
#redondearemos a un numero mayor, si es un 1.7 sera 2 y asi

nombres['matriculas']  = np.ceil(np.random.exponential(size=total_alumnos)).astype(int)
nombres.head()
```

```python
'''bien ya no tenemos valores en 0 pero siguen siendo numeros
muy bajitos como par ahacer nuestros calculos, asi que multiplicaremos
los valores por 1.5
'''
nombres['matriculas']  = np.ceil(np.random.exponential(size=total_alumnos)* 1.5).astype(int)
nombres.head()
```

# Seleccionando Alumnos

```python
nombres.head()
```

```python
'''
del df anterioro nos falta el id de los cursos q ya lo tenemos
en el df de cursos
'''
cursos.head()
```

```python
'''ahora quiero q hayan algunos cursos con mayor probabilidad
de ser elegidos q oros, crearemos probabilidad.
primero creare una lista vacia donde guardaremos el id y el indice del curso

crearemos una variable x donde invocremos al metodo random
del tamaño de mis cursos q en este caso son 13,
puedo llamar al tamañao del df con len.
ahora x ya es un array con 13 valores
ahora puedo elegir la probabilidad para cada valor de los 13
ahora a x le dare valores random, cuántos valores
random se deben crear? pues segun la cantidad de cursos
asi que a x le pasare el metodo random y la cantidad de cursos
en vez de decirle q 13 lo hare con (len(cursos))
ahora elegire la probabilidad para cada valor de los 13
se hace asi :cada valor x / en la suma de x
esto lo guardadre en una variable prob

'''

todas_matriculas =[]
x = np.random.rand(len(cursos))
prob = x / sum(x)
prob
```

```python
''' bien en la celda anterior prob tiene 13 valores
y en cada valor tengo la probabilidad de q ese curso tenga más
probabilidades de ser elegido q otro curso, de esta manera
la elección de curso sera mas avanzada '''
```

```python
#la idea es relacionar ambos df en uno para poner el id en el df nombres
'''
bien, haremos un recorrido dentro del df nombres, para identificar
cuantas matriculas tiene cada registro y elegiremos los cursos segun
la cantidad asi si un alumno iene 2 matriculas, aleatoriamente
le asignaremos dos cursos.

'''

'''
bien, haremos un for, identificaremos el indice y la fila (row) del df nombres
que es donde tenemos a los alumnos.
usaremos el iterrows par recorrer fila por fila el df
ahora identificaremos el id del alumno para ello
llamaremos la fila(row) id y le pedire q seleccioe la columna id_alumno
esto quedara en la variable id.
ahora enla variable matricula haremos lo mismo, llamamos
la fila(row) y le decimos q seleccione la columna matricula
'''

'''
ahora haremos un for anidado, en este for seleccionaremos
aleatoriamente la cantidad de cursos segun nel resultado anterior
segun matriculas, entonces
necesitamos una funcion q nos eliga o nos de valores
random, y que nos elija los valores esto lo haremos con choice
qué vamos a escojer¡? los cursos, los cursos estan en el df
cursos pero quiero que elija el index lo hago pasando como prametro
(cursos.index) asi elejira aleatoriamente el indice de algun curso
tantas veces como mi alumno este matriculado.
esto lo guardaremos en una variable q llamremos mat
necesito tambien el indice del alumno, asi que añadiremos id al inicio
adicionaremos tambien la probabilidad para q lo tome en
cuenta a la hora de hacer la selección, lego
añadimos el resultado a la lista vacia q habiamos creado
finalmente crearemos un df con esta info, nombramos las columnas
'''

for index, row in nombres.iterrows():
  id = row.id_alumno
  matriculas = row.matriculas
  for i in range(matriculas):
    mat= [id, np.random.choice(cursos.index, p = prob)]
    todas_matriculas.append(mat)

matriculas = pd.DataFrame(todas_matriculas, columns=['id_alumno', 'id_cursos'])
matriculas.head()
```

```python
'''
ahora con lo anterior podemos generar un reporte, como el q ya hicimos
ahoora mostraremos con el nombre del curso, cuántos alumnos
estan inscrios en cada curso.
Prinero haremos un group by del df que resulto q es matriculas, todo df puede
agruparse, lo haremos con 'id_curso', a continuación contaremos cuántos
alumnos estan matriculados en cada curso y luego haremos un join para
que se muestre el nombre del curso, para ello elegimos el df con el que lo
cruzaremos que es el que contiene los nombres de los cursos osea el df cursos
y elegiremos la columna q tiene esa información

'''

matriculas.groupby('id_cursos').count().join(cursos['nombre_curso'])
```

```python
#podemos mejorar el orden en que estan las columnas
 #y cambiar el nombre de las columnas
# y ordenamos de forma descendiente

matriculas.groupby('id_cursos').count().join(cursos['nombre_curso'])[['nombre_curso','id_alumno']].rename(columns={'id_alumno':'Total'}).sort_values('Total', ascending= False)
```

# Creando tablas en SQL
## Creando Bancos SQL

```python
'''
Grabaremos datos d enuestro df dentro de un banco sql
y leeros algunas querys sobre este banco
'''
```

```python
#para crear un banco necesitamos sqlalchemy
#pero solo importaremos algunas librerias
#create_engine =>  para crear el motor
#MetaData => paar trabajar con los datos
#Table => para crear la tabla

from sqlalchemy import create_engine, MetaData, Table, inspect, text
```

```python
matriculas_por_curso.head(3)
```

```python
#creamos el motor
'''para ello guardaremos el motor en una variable que llamaremos
engine, llamamos a la libreria q importamos en los parentesis pondremos el motor
hay varios tipos de motores por ejemplo postgresql - mysql - oracle
pero en este caso usaremos sqlite que viene en colab, asi q nos permite
trabajarlo en la nube.
el formato para crearlo es
://aqui va el camino donde est el banco de datos
si fuera por ejemplo localmente seria path y el nombre del banco.db
seria asi ('sqlite://path/nombre.db')
pero como estamos trabajando en la nube lo dejaremos en blanco
lo crearemos asi en memoria asi q mientras nuestro notebook este funcionando
el motor estara disponible, si quiero especificar que va  a
quedar en memoria ponemos /:memory:
'''

engine = create_engine('sqlite:///:memory:')
engine
```

```python
#ahora crearemos la tabla
#para guardarlo podemos usar cualquiera de los formatos que
#aprendimos para exportar ó cualquiero otro similar,
#entre parentesis ponemos el nombre que le queremos dar a
#la tabla y el motor q en este caso esta en la variable engine

matriculas_por_curso.to_sql('matriculas', engine)
```
