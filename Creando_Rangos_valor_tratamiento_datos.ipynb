```python
import pandas as pd
```

```python
datos = pd.read_csv('alquiler.csv',sep= ';')
datos.head(10)
```

```python
#agruparemos por cantidades de cuartos q tenga
#el inmueble y asi ver las frecuencias

# quiero agrupar
#1 y 2 cuartos
# 3 y 4
# 5 y 6
#7 ó más

'''
para ello crearemos condicionales para esta seleccion
como haciendo cortes, para obtener los registrps de
1 y 2 cuartos los limites serian desde 0 - 2 y asi.
'''

clases =  [0, 2, 4, 6, 100]
```

```python
#ahora llamremos cut e
#este metodo realizara los cortes
#le diremos q tome los datos de la columna Cuartos
#y q tome como base para acer los cortes la
#lista clases

cuartos = pd.cut(datos.Cuartos, clases)
cuartos
```

```python
'''
bien aqui genero una series, de categorias, asi q
recorre el df y en seguida le pone a qué categoria
pertenece

bien aqui hay inmuebles q tienen 0 cuartos y
en nuestras categorias el 0 no es considerado
asi  q lo deja nulo

veamos las frecuencias
'''
```

```python
pd.value_counts(cuartos)
```

```python
#generaremos nuestras etiquetas

label = ['1 y 2 cuartos', '3 y 4 cuartos', '5 y 6 cuartos', '7 ó más cuartos' ]
```

```python
#bien ahora al llmar al metodo llmare
#o le pondre las etiquetas

cuartos = pd.cut(datos.Cuartos, clases, labels = label )
```

```python
pd.value_counts(cuartos)
```

```python
#vamos a incluir los valores de 0 cuartos
#dentro de la primer categoria

cuartos = pd.cut(datos.Cuartos, clases, labels = label, include_lowest= True )
cuartos
```
