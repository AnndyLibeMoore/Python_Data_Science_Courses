```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (15, 10))
```

```python
datos = pd.read_csv('/content/alquiler_residencial_limpio.csv', sep = ';')
datos.head(10)
```

```python
#creamos el area del gráfico
area = plt.figure()
```

```python
#las dimensiones
#generare 4 graficos, dos lineas, 2 columnas
#primer grafico g1, le dire q se ponga dentro
#de area y adicione un gráfico,
#especifico q quiero 2 lineas y 2 columnas
#y que estara en la primer posicion (2,2,1)

g1 = area.add_subplot(2,2,1)
g2 = area.add_subplot(2,2,2)
g3 = area.add_subplot(2,2,3)
g4 = area.add_subplot(2,2,4)
```

```python
#Ahora configuramos los tipos de graficos

#bien, ahora los gráficos
#en el g1 quiero un grafico de dispersion
#usa 2 columnas de medicion,
#este grafico se llama scatter
#debo elegir las dos columnas q usará
g1.scatter(datos.Valor, datos.Area)
g1.set_title('Valor por Área')

#para el segundo grafico quiero un histograma
#con la columna valor
g2.hist(datos.Valor)
g2.set_title('Histograma')

#para el g3 sera un gráfico de lineas un plot
#generare una muestra de datos
#esto lo q hará es seleccionar lineas aleatorias
#de mi df esto lo hago con sample
#debo elegir la columna y la cantidad de registros
#tambien rehacer los indices para evitar problemas
datos_g3 = datos.Valor.sample(100)
datos_g3.index = range(datos_g3.shape[0])
g3.plot(datos_g3)
g3.set_title('Muestra (Valor)')

#para g4 un codigo de barras bar
#agreamos un objeto grupo q va a agrupar por tipo
#y traera el valor
#ahora las etiquetas de las barras
#generare la media del objeto grupo y mostrare
#los indices
#para los valores sobre cada barra creare
#el objeto valores pero envez del indice
#aqui llevara los valores
#y ahora ya paso los datos para
#generar el grafico
grupo = datos.groupby('Tipo')['Valor']
label = grupo.mean().index
valores = grupo.mean().values
g4.bar(label, valores)
g4.set_title('Valor Medio por Tipo')
```

```python
#ahora mostramos los graficos
#llamando a area
area
```

```python
#ahora quiero poner estas imagenes en el documento
#podemos exportar esto a colab con el metodo
#savefig, con dpi especificamos los pixeles
#ajustamos los bordes con bbox_inches, tight
#hace q los bordes aparezcan ajustado sy no se vean
#los bordes blancos

area.savefig('grafico.png', dpi = 300, bbox_inches = 'tight')
```
